# Matching practice

Reset index of mentor/mentee dataframe

```python
if df_mentee.index.name != 'mentee_id':
  df_mentee.reset_index(inplace=True)
  df_mentee.set_index('mentee_id', drop=True, inplace=True)

if df_mentor.index.name != 'mentor_id':
  df_mentor.reset_index(inplace=True)
  df_mentor.set_index('mentor_id', drop=True, inplace=True)
  df_mentor.sort_index(inplace=True)

df_mentor['mentee_MSc'] = ''
df_mentor['mentee_PhD'] = ''

df_mentor['no_match_MSc'] = 0
df_mentor['no_match_PhD'] = 0
```

## create a copy of "mentee order of mentor"
create a copy of mentor response or 'wish list of mentors' in a dictionary

both mentors has responded (based on the the mentors' response) and not responded mentors (based on the order from mentees)

NOTE: some mentors put both MSc/PhD mentee (the mentee considers to apply both MSc and PhD programmes) in MSc list, however, to simplify the matching process, those mentees are classified as PhD mentees

```python

import ast

di_mentor_response = {}
di_mentor_response['MSc'] = {}
di_mentor_response['PhD'] = {}

for mentor_id in df_mentor.index:

  ## if mentor has response (select mentees), then copy the order of the mentee
  if df_mentor.loc[mentor_id, 'mentee_response']:
    if df_mentor.loc[mentor_id, 'response_MSc'] != '[]':
      li_msc = ast.literal_eval( df_mentor.loc[mentor_id, 'response_MSc'] )
      # remove PhD mentee from MSc list
      for mentee_id in li_msc:
        if mentee_id[0] == 'P': li_msc.remove(mentee_id)
      if li_msc != []:
        di_mentor_response['MSc'][mentor_id] = li_msc

    if df_mentor.loc[mentor_id, 'response_PhD'] != '[]':
      di_mentor_response['PhD'][mentor_id] = ast.literal_eval( df_mentor.loc[mentor_id, 'response_PhD'] )

  ## if mentor has not selecting mentees, then copy the order from mentee order
  else:
      for degree in ['MSc', 'PhD']:
        li_mentee = []
        for i in range(1,6):
            col = f'{degree}: no_{i}'
            if df_mentor.loc[mentor_id, col] != '': li_mentee += ast.literal_eval( df_mentor.loc[mentor_id, col] )
        if len(li_mentee) > 0:
          di_mentor_response[degree][mentor_id] = li_mentee
```

```python
def select_the_mentor(di_mentee, list_mentors):
  '''return the mentor_id in list_mentors that has highest order'''

  for i in range(1, 6):
    if di_mentee[f'no_{i}'] in list_mentors:
      return di_mentee[f'no_{i}']

  print('Error! Cannot find mentor!')
```

```python
def show_match_sta(df_mentee ):

  # print out some statistics
  df=df_mentee.loc[df_mentee['degree']==degree]
  print('total {} mentee {}'.format(degree, len(df)))
  df=df.loc[df['mentor']=='']
  print('{} of {} mentees have not match to a mentor.'.format(len(df), degree))
```

```python
## find the top choice in the copy of response list (di_mentor_response)

def run_matching(select_mentor, di_mentor_response, df_mentee, df_mentor, degree, run):
  '''to find the pair of mentor/mentee in select_mentor list, based on the
     orders in di_mentor_response list'''

  di_match = {}

  ### this code is to generate `di_match[mentee_id]=list(mentor_id)`
  # which stores the list of mentors appointed to the mentee

  for mentor_id in di_mentor_response[degree].keys():

    # skip if not in the select list
    if mentor_id not in select_mentor: continue

    while len(di_mentor_response[degree][mentor_id]) > 0:

      mentee_id = di_mentor_response[degree][mentor_id][0]
      if df_mentee.loc[mentee_id, 'mentor'] != '':
          di_mentor_response[degree][mentor_id].remove(mentee_id)

      else:
          if mentee_id not in di_match.keys(): di_match[ mentee_id ] = []
          di_match[ mentee_id ].append( mentor_id )
          break


  ### based on di_match to find the final match pair
  count=0
  for mentee_id in di_match.keys():

    # if only one mentor appointed, matched!
    if len(di_match[mentee_id]) == 1:
      mentor_id = di_match[mentee_id][0]
      match_note = f'single match, run {run}'

    # if more than one mentor appointed
    else:
      # select the mentor by the order from mentee
      mentor_id = select_the_mentor(df_mentee.loc[mentee_id], di_match[mentee_id])
      match_note = 'select from {}, run {}'.format(len(di_match[mentee_id]), run)

    print(f'mentee {mentee_id} is matched to mentor {mentor_id} by {match_note}')

    # store match data
    df_mentee.loc[mentee_id, 'mentor'] = mentor_id
    df_mentee.loc[mentee_id, 'match_note'] = match_note

    df_mentor.loc[mentor_id, f'mentee_{degree}'] += mentee_id +' '
    df_mentor.loc[mentor_id, f'no_match_{degree}'] += 1

    select_mentor.remove(mentor_id)

    count += 1

  print(f'{count} of mentees are matched in this run.')
  return select_mentor, di_mentor_response, df_mentee, df_mentor, count
```

function to run multiples search runs

```python
def run_matching_runs(select_mentor, degree, di_mentor_response, df_mentee, df_mentor):

  run = 1

  while True:
    print('\n{} mentors to match at run {}'.format(len(select_mentor), run))

    select_mentor, di_mentor_response, df_mentee, df_mentor, count = run_matching(select_mentor, di_mentor_response, df_mentee, df_mentor, degree, run)
    run += 1

    show_match_sta(df_mentee)
    if count == 0 or run== 10: break
```

to remove the mentor that reachs max capacity

```python
def check_capacity(select_mentor, degree, df_mentor):
  '''remove the mentor who reaches max capacity from the selecting list'''

  select_mentor_ret = select_mentor.copy()
  for mentor_id in select_mentor:
    no_match = df_mentor.loc[mentor_id, f'no_match_{degree}']
    no_capacity = df_mentor.loc[mentor_id, f'capacity_{degree}']
    # print(f'{mentor_id} no match: {no_match}, {degree} capacity: {no_capacity}')
    if no_match > no_capacity:
      print(f'warning! Mentor {mentor_id} is over {degree} capacity {no_capacity}')

    if no_match == no_capacity:
      print(f'{mentor_id} reaches {degree} capacity {no_capacity}')
      select_mentor_ret.remove(mentor_id)

  return select_mentor_ret
```
